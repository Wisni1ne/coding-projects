```python
import pandas
import numpy as np
import re
import nltk
import tensorflow as tf
import random
import csv
import tkinter as tk
from tkinter import *
from tkinter import filedialog, Text
import os
from PIL import ImageTk,Image
import pygame
from random import *
```

```python
#Displays next line of code
def nextLine():
    x = randint(1, len(csv_arr)-1)
    next_Sentence = csv_arr[x]
    CSV_Display = tk.Label(text=next_Sentence[0], relief=tk.RIDGE, width=100, height=2)
    CSV_Display.grid(row=2, column=1)
    prediction = tk.Label(text=next_Sentence[1], relief=tk.RIDGE, width=15, height=2)
    prediction.grid(row=2, column=2)
    Actual = tk.Label(text=next_Sentence[2], relief=tk.RIDGE, width=15, height=2)
    Actual.grid(row=2, column=3)
    print(csv_arr)
        
```

```python
#Picks the Language
def Language():
    
    if tkvar.get() == 'English':
        
        pygame.mixer.init(44100, -16,2,2048)
        pygame.mixer.music.load('TWSSMS.mp3') #Loading File Into Mixer
        pygame.mixer.music.play(0) #Playing It In The Whole Device
                
    if tkvar.get() == 'Spanish':
        
        pygame.mixer.init(44100, -16,2,2048)
        pygame.mixer.music.load("mp3-converts/twss_spa.mp3") #Loading File Into Mixer
        pygame.mixer.music.play(0) #Playing It In The Whole Device
        
    if tkvar.get() == 'German':
        
        pygame.mixer.init(44100, -16,2,2048)
        pygame.mixer.music.load("mp3-converts/twss_ger.mp3") #Loading File Into Mixer
        pygame.mixer.music.play(0) #Playing It In The Whole Device
        
    if tkvar.get() == 'Polish':
        
        pygame.mixer.init(44100, -16,2,2048)
        pygame.mixer.music.load("mp3-converts/twss_pol.mp3") #Loading File Into Mixer
        pygame.mixer.music.play(0) #Playing It In The Whole Device
        
    if tkvar.get() == 'Japanese':
        
        pygame.mixer.init(44100, -16,2,2048)
        pygame.mixer.music.load("mp3-converts/twss_jap.mp3") #Loading File Into Mixer
        pygame.mixer.music.play(0) #Playing It In The Whole Device
```

```python
#---------------MAIN UI--------------------

root = tk.Tk()

root.title('Michael Bot')

csv_arr = [] 
with open("classifier_results.csv", newline='') as f:
    reader = csv.reader(f)
    for line in reader:
        csv_arr.append(line)

# Create Tkinter variable.
tkvar = StringVar(root)

# Dictionary with language options.
choices = [ 'English','Spanish','German','Polish','Japanese', 'Off' ]

#Displays the Sentence
CSV_Display = tk.Label(text='Sentence Display', relief=tk.RIDGE, width=100, height=2)
CSV_Display.grid(row=2, column=1)

#Displays the prediction
prediction = tk.Label(text='Prediction', relief=tk.RIDGE, width=15, height=2)
prediction.grid(row=2, column=2)

#Displays the Actual
Actual = tk.Label(text='Actual', relief=tk.RIDGE, width=15, height=2)
Actual.grid(row=2, column=3)

#Next Button
next_ = tk.Button(root, text='Next', font=40, height=2, command=nextLine)
next_.grid(row=2, column=4)

#Entry Field
entry = tk.Entry(root, font=40, width=100)
entry.grid(row=3, column=1)

#Displays Prediction for user
prediction2 = tk.Label(text='Prediction', relief=tk.RIDGE, width=15, height=2)
prediction2.grid(row=3, column=2)

#Submit Button
submit = tk.Button(root, text='Submit', font=40, height=2, command=Language)
submit.grid(row=3, column=3)

# Create dropdown box for a list of language options.
popupMenu = OptionMenu(root, tkvar, choices[1], *choices)
Label(root, text="Language:").grid(row = 4, column = 2)
popupMenu.grid(row = 4, column =3)

root.mainloop()
```
